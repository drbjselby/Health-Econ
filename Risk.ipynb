## A person's health in the future is inherently a risky business

Let's call $p$ the probability of being seriously injured

```python
p = 0.50 #the probability of injury
```

And perhaps if you get injured, you can earn about $10,000 in workers compensation benefits. 

If you are not injured then you earn your regular salary of $40,000

Then, on average you can earn an expected income of: 

```python
EI = p*10000 + (1-p)*40000

print("EI = $" + str(round(EI,2)))
```

What happens to your expected income as p increases?

Let's call expected income
$$E(I) =p*10000 + (1-p)*40000 $$

So as the probability of injury increases then the person will have a lower average income in this risky world.

```python
p = np.linspace(0,1)
EI = p*10000 + (1-p)*40000 
plt.plot(p,EI)
plt.title('Expected Income')
plt.ylabel('Exp. Income')
plt.xlabel('p')
plt.grid(axis='y', alpha=0.75)
```

### Utility of Income

We now must consider how much that a person enjoys an extra dollar of income:
* Does a $1 matter more when income is higher or lower?

To answer this, we need to create a Utility of Income function
$$U(I)$$

#### Risk Aversion: this person would prefer a certain outcome of $E(I)$ over an uncertain one:
$$ U[E(I)] > E[U]$$

When does this happen?

Imagine that our utility function has decreasing marginal utility: $$\frac{dU}{dI} > 0 \quad \quad \frac{d^2U}{dI^2} < 0$$

* Example: $U(I) = I^{0.5}$

```python
I = np.linspace(0,50000)
U =  np.power(I,0.5)
plt.plot(I, U) 
plt.title('Utility of Income')
plt.ylabel('Utility')
plt.xlabel('Income')
plt.grid(axis='y', alpha=0.75)
```

If we can expect 10,000 or 40,000 with certainty, our utility would be, respectively
$$U(10000) = 100$$
$$U(40000) = 200$$

If we got the expected value of income with certainty, we would get utility:
$$U(EI) = U[p*10000 + (1-p)*40000]$$

Moreover, the expected utility, if we take the gamble, is
$$E(U) = pU(10000) + (1-p)U(40000)$$

```python
I1 = 10000
I2 = 40000
u1 = np.power(I1,0.5)
u2 = np.power(I2,0.5)
p= 0.50
EI = p*10000 + (1-p)*40000 
UEI = np.power(EI,0.5)
EU = p*u1 + (1-p)*u2

plt.plot(I, U) 
plt.plot((I1,I2), (u1, u2), 'r-')
plt.plot((EI,EI), (0, UEI ), 'm-.')
plt.plot((EI,0), (EU,EU), 'k--')
plt.plot((EI,0), (UEI,UEI), 'k--')

plt.title('Utility of Income')
plt.ylabel('Utility')
plt.xlabel('Income')
plt.axis([5000,45000,90,250])

plt.grid(axis='y', alpha=0.75)
```

The top black dashed line represents the utility of getting the expected value with 100% certainty. The black bottom dashed line represents the expected utility of the gamble. This utility function represents the risk averse person.

#### Risk Neutral : this person is indifferent between a certain outcome of $E(I)$ over an uncertain one:
$$ U[E(I)] = E[U]$$

When does this happen?

Imagine that our utility function has constant marginal utility: $$\frac{dU}{dI} > 0 \quad \quad \frac{d^2U}{dI^2} = 0$$

* Example: $U(I) = I$


```python
I = np.linspace(0,50000)
U =  np.power(I,1)
plt.plot(I, U) 
plt.title('Utility of Income')
plt.ylabel('Utility')
plt.xlabel('Income')
plt.grid(axis='y', alpha=0.75)
```

```python
I1 = 10000
I2 = 40000
u1 = np.power(I1,1)
u2 = np.power(I2,1)
p= 0.50
EI = p*10000 + (1-p)*40000 
UEI = np.power(EI,1)
EU = p*u1 + (1-p)*u2

plt.plot(I, U) 
plt.plot((I1,I2), (u1, u2), 'r-')
plt.plot((EI,EI), (0, UEI ), 'm-.')
plt.plot((EI,0), (EU,EU), 'k--')
plt.plot((EI,0), (UEI,UEI), 'k--')

plt.title('Utility of Income')
plt.ylabel('Utility')
plt.xlabel('Income')
plt.axis([5000,45000,5000,45000])

plt.grid(axis='y', alpha=0.75)
```

The  black dashed line represents the utility of getting the expected value with 100% certainty, which is equal to expected utility of the gamble. This utility function represents the risk neutral person.

#### Risk Seeking : this person is prefers a gamble over a certain outcome of $E(I)$:
$$ U[E(I)] < E[U]$$

When does this happen?

Imagine that our utility function has constant marginal utility: $$\frac{dU}{dI} > 0 \quad \quad \frac{d^2U}{dI^2} > 0$$

* Example: $U(I) = I^2$


```python
I = np.linspace(0,50)
U =  np.power(I,2)
plt.plot(I, U) 
plt.title('Utility of Income')
plt.ylabel('Utility')
plt.xlabel('Income ($1000s)')
plt.grid(axis='y', alpha=0.75)
```

```python
I1 = 10
I2 = 40
u1 = np.power(I1,2)
u2 = np.power(I2,2)
p= 0.50
EI = p*10 + (1-p)*40
UEI = np.power(EI,2)
EU = p*u1 + (1-p)*u2

plt.plot(I, U) 
plt.plot((I1,I2), (u1, u2), 'r-')
plt.plot((EI,EI), (0, EU ), 'm-.')
plt.plot((EI,0), (EU,EU), 'k--')
plt.plot((EI,0), (UEI,UEI), 'k--')

plt.title('Utility of Income')
plt.ylabel('Utility')
plt.xlabel('Income ($1000s)')
plt.axis([0,55,0,2500])

plt.grid(axis='y', alpha=0.75)
```

Now, the bottom black dashed line represents the utility of getting the expected value with 100% certainty. The black top dashed line represents the expected utility of the gamble. This utility function represents the risk seeking person.

# Insurance

Now, many people buy "insurance" in health. Effectively, they are insuring against the expected loss of income from being sick. So how does this work?

Let's consider a very simple insurance plan. This consists of the following:
* q = coverage amount, the amount that an insurer will pay out in the event that a claim is made
* a = premium rate, the amount that a person will pay as a fraction of the total covered amount ($aq$ is the insurance premium payment)

When the person is healthy:
* $I_H'$ = income when healthy with insurance = $I_H - aq$
When the person is sick: 
* $I_S'$ = income when sick with insurance = $I_S - aq+ q$

The **actuarially fair rate of insurance** is the $a$ that satisfies
$$I_H' =I_S'$$
so that they can have the same expected income in either state.
We will assume that the person selects to insure for the entire difference in income from healthy to sick:
$$q=I_H-I_S$$

The person gets sick with probability $p$, so the expected diffence between income when healthy and sick is $pq$. Thus it must be the case that the actuarially fair rate is 
$$a=p$$

## Taking this back to our utility:
If a person makes 40,000 when healthy and gets sick with probability $p=0.5$, then there is a 50% chance of making 10,000. If they insure for the entire loss, then they will pay a premium
$$0.5*30000=15000$$ 

Thus their income after insurance when healthy is 
$$I_H' = 40000-15000=25000$$

If they get sick, then their income after insurance is
$$I_S' = 10000-15000+30000=25000$$

Moreover, this is the expected income with a probability of sickness of 0.5!
$$E(I) = 0.5*10000 + (1-0.5)*40000 =25000$$

Thus, by buying insurance, this person can **guarantee** an income of 25,000. Then, we can say:
1. The risk-averse person would prefer the insurance. They prefer to get 25,000 with certainty than a gamble
2. The risk-neutral person would not have a strict preference for insurance
3. The risk-seeking person would not buy insurance because the guaranteed income gives less satisfaction than the gamble
