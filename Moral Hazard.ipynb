# Moral Hazard

**Definition** Information Asymmetry
* A situation where agents in a market have an unequal amount of information about the transaction (e.g. quality)

**Definition** Moral Hazard
* A situation where an agent in a market enters changes their behavior because of the exisitance of insurance against loss
    - Example: Kathy goes to the doctor once a year for regular checkups. She now receives insurance which covers a variety of services, including massage. Although she wouldn't buy a massage without insurance, she now wants it with insurance. 
    
## Ex Ante Moral Hazard: 
Behavior changes that increase the likehood of needing to file an insurance claim
* Example: Eating more fatty foods, driving fast, etc.

## Ex Post Moral Hazard:
Behavior changes that occur after an incident occurs where insured individuals opt for more expensive remedies 
* Example: Opting for surgical instead of medical procedures 
![All the X Meme](https://i.imgflip.com/2xk20t.jpg)

## Pattern of Behavior:
1. Individual might experience bad thing
2. Individual insures against bad thing, reducing his price to pay for bad thing
3. Individual changes behavior that increases chances of the bad thing (ex ante) or increases the costs of recovering from bad thing (ex post)
4. Insurers cannot observe changes in behavior due to information asymmetry
5. Individuals risky behavior leads to social welfare loss 



```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(0,100)
Y = 100-X
fig, ax = plt.subplots()
plt.plot(X,Y, color='r')
plt.xlabel("Quantity of Services")
plt.ylabel("Price of Services")
plt.plot((20,20,0),(0,80,80), 'r--')
plt.text(20,80,s="Uninsured")
plt.text(60,40,s="Insured")
plt.plot((60,60,0),(0,40,40), 'b--')
plt.text(90,15,s="Demand")
plt.title("Social Loss due to Moral Hazard")
plt.xticks(np.arange(0, 110, step=10.0))
plt.axis([0,105,0,105])

plt.show()
```

The uninsured individuals pick **low quantity of services** due to their higher out of pocket price.

The insured individuals select **higher quantity of services** due to their lower out of pocket price.

# Bringing it to Health Insurance

* $X_i$ is the expected health expenditures in a year
* An insurance company offers a single policy with premium $P$ which covers all health expenditures in a year
* Customers are risk neutral, they will purchase insurance if and only if $$P<E[H_I]$$
* Insurers cannot discriminate against healthy and sick individuals 
* Health care costs are distributed
$$H_i \sim Uniform(0,\bar{H})$$
where $\bar{X}$ is the maximum that a person can expect to pay for healthcare. 


Now, customers will only buy insurance at a price of $P$ if they expect to have $E[H_I] > P$.

For $H_i\in (P,\bar{H})$, the customers would buy insurance:

```python
ub = 20000
lb = 0
p=(1/(ub-lb))

H = np.random.uniform(low=lb, high=ub, size=1000)
count, bins, ignored = plt.hist(H, 10, density=True,rwidth=0.75)
plt.plot(bins, (1/(ub-lb))*np.ones_like(bins), linewidth=2, color='r')
plt.xlabel("H")
plt.ylabel("Proportion")
plt.title("Sample from Uniform("+str(lb)+","+str(ub)+")")
plt.xticks(np.arange(0, ub+1000, step=1000.0), rotation=45)
plt.show()

hbar = 10000
x = [hbar, hbar, ub, ub]
y = [0, p, p, 0]
fig, ax = plt.subplots()
plt.plot((lb,lb),(0,p),'r--' )
plt.plot((ub,ub),(0,(1/(ub-lb))),'r--' )
plt.plot((hbar,hbar),(0,(1/(ub-lb))),'b--' )
plt.plot(bins, p*np.ones_like(bins), linewidth=2, color='r')
ax.fill(x, y,color='g', alpha=0.5)
plt.xlabel("H")
plt.ylabel("Probability")
plt.title("Uniform("+str(lb)+","+str(ub)+")")
plt.xticks(np.arange(0, ub+1000, step=1000.0), rotation=45)
plt.axis([lb-1000,ub+1000,0,p+0.00001])
plt.show()
```

### Only the sickest buy insurance!
This leads to an expected loss of 
$$E(Premium) - E(Payout) = \frac{10000+20000}{2} -10000 = -5000$$ 

![This or That Meme](https://i.imgflip.com/2xjtul.jpg) 

### Why not just raise price? 
Well, if price goes up, only the sickest (ones wanting to buy health insurance at such a high price) will be left


```python
hbar = 15000
x = [hbar, hbar, ub, ub]
y = [0, p, p, 0]
fig, ax = plt.subplots()
plt.plot((lb,lb),(0,p),'r--' )
plt.plot((ub,ub),(0,(1/(ub-lb))),'r--' )
plt.plot((hbar,hbar),(0,(1/(ub-lb))),'b--' )
plt.plot(bins, p*np.ones_like(bins), linewidth=2, color='r')
ax.fill(x, y,color='g', alpha=0.5)
plt.xlabel("H")
plt.ylabel("Proportion")
plt.title("Sample from Uniform("+str(lb)+","+str(ub)+")")
plt.xticks(np.arange(0, ub+1000, step=1000.0), rotation=45)
plt.axis([lb-1000,ub+1000,0,p+0.00001])
plt.show()
```

Now the expected loss is:
$$E(Premium) - E(Payout) = \frac{15000+20000}{2} -15000 = -2500$$ 

It is better! Keep raising prices! 

In the end, the company ends up only with the sickest individuals paying the full price of health care.

This is known as a
